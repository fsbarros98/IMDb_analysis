# Analyzing the first 2 genres

```r
genres=merged[,c(6,7)] #Just 2 genres
```

```r
length(unique(interaction(genres))) #338 unique combinations
```

```r
genres=table(interaction(genres)) #Creating a table with the interactions between genres
genres=melt(genres, value.name="count") #Converting the table into a dataframe
genres=subset(genres,count>0) #Some combinations of genres have 0 movies, which are useless
genres=separate(genres,'Var1',sep='[.]',into=c('genre1','genre2')) #Separating the Var1 column into 2 columns
```

```r
head(genres)
```

```r
#Professsor's code - Notebook 10
ggplot(genres, aes(x=genre1, y=genre2)) +
    geom_point(aes(size=count), shape=21, color="black", fill="cornsilk") +
    scale_size_area(max_size=10, guide=FALSE) +
    theme(axis.text.x = element_text(angle = 90))
```

# Taking the ratings into account

```r
genres_ratings=subset(merged,averageRating>0) #Movies with ratings
```

```r
genres_ratings=genres_ratings[,c(6,7,9)] #Only the first two genres and the ratings
```

```r
length(unique(interaction(genres_ratings[,c(1,2)]))) #Not all combinations from before have ratings...
```

```r
genres2=table(interaction(genres_ratings[,c(1,2)])) #Creating a table with the interactions between genres
genres2=melt(genres2, value.name="count") #Converting the table into a dataframe
genres2=subset(genres2,count>0) #Some combinations of genres have 0 movies, which are useless
genres2=separate(genres2,'Var1',sep='[.]',into=c('genre1','genre2')) #Separating the Var1 column into 2 columns
```

```r
#I didn't come up with a better idea to do this without a for cicle, but it's fast... Don't judge me xD
means=c()
a=seq(1,nrow(genres2),1)

for (i in a){
    means=rbind(means,mean(genres_ratings[genres_ratings['genre1']==genres2[i,1] & genres_ratings['genre2']==genres2[i,2],]$averageRating))
} 
```

```r
genres2$Avg_Rating=c(means)
```

```r
#Professsor's code - Notebook 10
ggplot(genres2, aes(x=genre1, y=genre2)) +
    geom_point(aes(size=Avg_Rating^2), shape=21, color="black", fill="cornsilk") +
    scale_radius(breaks=10) +
    theme(axis.text.x = element_text(angle = 90))
```
