```r
library(ggplot2)
library(plyr)
```

```r
title_akas=read.delim('/home/brunovaz19/Desktop/Data Science/Masters/Year 1/Semester 1/Visualização de Dados/IMDb_Datasets_Filtered/cleaned_filtered_title_akas.tsv')
```

```r
options(repr.plot.width=9,repr.plot.height=15)
```

```r
freq=count(title_akas,'region') #frequency of regions
```

```r
lengths(freq) #237 regions
```

```r
freq_0_100=freq[(freq$freq>=0) & (freq$freq<=100),]
freq_100_1000=freq[(freq$freq>100) & (freq$freq<=1000),]
freq_1000_10000=freq[(freq$freq>1000) & (freq$freq<=10000),]
freq_above_10000=freq[freq$freq>10000,]
```

```r
lengths(freq_0_100)
lengths(freq_100_1000)
lengths(freq_1000_10000)
lengths(freq_above_10000)
117+49+36+35 == 237 #Great
```

```r
ggplot(freq_100_1000,aes(x=freq,y=reorder(region,freq)))+
geom_bar(stat='identity',color=1,fill=4,alpha=.3) +
labs(x='Frequency',y='Region') +
scale_x_continuous(breaks=c(seq(0,1000,100)))+
theme(text=element_text(size=15))
```

```r
ggplot(freq_1000_10000,aes(x=freq,y=reorder(region,freq)))+
geom_bar(stat='identity',color=1,fill=4,alpha=.3) +
labs(x='Frequency',y='Region') +
scale_x_continuous(breaks=c(seq(0,10000,1000)))+
theme(text=element_text(size=15))
```

```r
ggplot(freq_above_10000,aes(x=freq,y=reorder(region,freq)))+
geom_bar(stat='identity',color=1,fill=4,alpha=.3) +
labs(x='Frequency',y='Region') +
scale_x_continuous(breaks=c(seq(0,410000,100000)))+
theme(text=element_text(size=15))
```
